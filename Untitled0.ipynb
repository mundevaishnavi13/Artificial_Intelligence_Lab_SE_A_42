The code for to find inverse of matrix with using the function

```python
import numpy as np

# Define the 3x3 matrix as a list of lists
matrix_list = [
    [2, 1, 1],
    [1, 1, 1],
    [1, 2, 1]
]

# Convert the list of lists to a NumPy array
matrix_np = np.array(matrix_list)

# Display the original matrix
print("Original Matrix:")
print(matrix_np)

# Calculate the determinant to check if the inverse exists
determinant = np.linalg.det(matrix_np)

print(f"\nDeterminant of the matrix: {determinant}")

# Check if the determinant is close to zero (indicating no inverse or singular matrix)
if abs(determinant) < 1e-9: # Using a small threshold for floating point comparison
    print("\nInverse does not exist because the determinant is zero (or very close to zero).")
else:
    # Calculate the inverse using NumPy's built-in function
    inverse_matrix = np.linalg.inv(matrix_np)

    # Print the inverse matrix
    print("\nInverse Matrix:")
    print(inverse_matrix)
```
