```python
# Filter the DataFrame for 'Iris-setosa'
setosa_data = df[df['Species'] == 'Iris-setosa']

# Compute aggregate statistics for 'Iris-setosa'
agg_stats_setosa = setosa_data.agg({
    'SepalLengthCm': ['sum', 'mean', 'count', 'min', 'max'],
    'SepalWidthCm': ['sum', 'mean', 'count', 'min', 'max'],
    'PetalLengthCm': ['sum', 'mean', 'count', 'min', 'max'],
    'PetalWidthCm': ['sum', 'mean', 'count', 'min', 'max']
})

print("\nAggregate Statistics for Iris-setosa:\n")
print(agg_stats_setosa)
```

```python
# Filter the DataFrame for 'Iris-versicolor'
versicolor_data = df[df['Species'] == 'Iris-versicolor']

# Compute aggregate statistics for 'Iris-versicolor'
agg_stats_versicolor = versicolor_data.agg({
    'SepalLengthCm': ['sum', 'mean', 'count', 'min', 'max'],
    'SepalWidthCm': ['sum', 'mean', 'count', 'min', 'max'],
    'PetalLengthCm': ['sum', 'mean', 'count', 'min', 'max'],
    'PetalWidthCm': ['sum', 'mean', 'count', 'min', 'max']
})

print("\nAggregate Statistics for Iris-versicolor:\n")
print(agg_stats_versicolor)
```
