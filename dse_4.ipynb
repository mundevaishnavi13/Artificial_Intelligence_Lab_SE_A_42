```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Load dataset - changed to an existing file in Colab sample data
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Show dataset in output (first few rows)
print("Dataset Head:")
print(df.head())

# Select numeric column - using 'housing_median_age'
plot_data = df['housing_median_age']

# Mean and Standard Deviation
mean = np.mean(plot_data)
std = np.std(plot_data)

# Histogram (Frequency)
plt.hist(plot_data, bins=30, density=True, alpha=0.6, color='g') # Using density=True for comparison with PDF
x = np.linspace(min(plot_data), max(plot_data), 100)
y = norm.pdf(x, mean, std)
plt.plot(x, y, 'r-', lw=2)

plt.title("Normal Distribution Histogram of Housing Median Age")
plt.xlabel("Housing Median Age")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
data = df['housing_median_age']
print("Values used from dataset (first 5 rows):")
print(data.head())
mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std_dev = np.std(data)
print("\nMean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
import random
n = int(input("Enter number of coin tosses: "))

heads = 0
tails = 0
for i in range(n):
    toss = random.choice(['Heads', 'Tails'])

    if toss == 'Heads':
        heads += 1
    else:
        tails += 1
# Calculate probabilities
prob_heads = heads / n
prob_tails = tails / n
print("Number of Heads:", heads)
print("Number of Tails:", tails)

print("Probability of Heads:", prob_heads)
print("Probability of Tails:", prob_tails)
```
