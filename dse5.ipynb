```python
import pandas as pd

# Load dataset using the correct filename
df = pd.read_csv("student_math_clean.csv")

# Display original dataset
print("Original Dataset:")
display(df.head())

# Filtering condition using correct column names and condition for study_time
filtered_data = df[(df['age'] > 18) & (df['study_time'] == '>10 hours')]

# Display filtered dataset
print("\nFiltered Dataset (Age > 18 and StudyHours > 10):")
display(filtered_data)
```

```python
import pandas as pd

# Load dataset using the correct filename
df = pd.read_csv("student_math_clean.csv")

# Display dataset
print("Student Dataset:")
display(df.head())

# Aggregation functions (only for numerical columns)
aggregation = df.agg({
    'age': ['count', 'mean', 'min', 'max', 'sum']
})

print("\nAggregation Results:")
print(aggregation)
```
